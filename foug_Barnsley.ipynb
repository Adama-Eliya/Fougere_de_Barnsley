```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 100000
x,y = 0,0
x_points = []
y_points = []
cpt = 0
while cpt < n:
    r = np.random.rand()
    if r < 0.01:
        x,y = 0, 0.16*y
    elif r < 0.86:
        x,y = 0.85*x + 0.04*y, -0.04*x + 0.85*y + 1.6
    elif r < 0.93:
        x,y = 0.2*x - 0.26*y, 0.23*x + 0.22*y + 1.6
    else:
        x,y = -0.15*x + 0.28*y, 0.26*x + 0.26*y + 0.44
    x_points.append(x)
    y_points.append(y)
    cpt +=1

plt.style.use("dark_background")
plt.figure(figsize = (6,10))
plt.scatter(x_points, y_points, s= 0.1,c= range(n), cmap = "Reds",alpha = 0.5)
plt.axis("off")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def generate_fern(n_points):
    points = np.zeros((n_points,2))
    x,y = 0,0
    for i in range(n_points):
        r = np.random.rand()
        if r < 0.01:
            x,y = 0, 0.16*y
        elif r < 0.86:
            x,y = 0.85*x + 0.04*y, -0.04*x + 0.85*y + 1.6
        elif r < 0.93:
            x,y = 0.2*x - 0.26*y, 0.23*x + 0.22*y + 1.6
        else:
            x,y = -0.15*x + 0.28*y, 0.26*x + 0.26*y + 0.44
        points[i] = [x,y]
    return points
np_points = 10000
points = generate_fern(n_points)

    
fig,ax = plt.subplots()
ax.set_xlim(-3,3)
ax.set_ylim(0,10)
ax.set_aspect("equal")
ax.axis("off")

scatter = ax.scatter(points[:,0],points[:,1], s= 0.1, color = 'green')

def update(frame):
    alpha = frame / 100
    scatter.set_facecolor((0,alpha,0))
    return scatter,
ani = animation.FuncAnimation(fig,update,frames= 100,blit = True)
plt.show()
    
```
